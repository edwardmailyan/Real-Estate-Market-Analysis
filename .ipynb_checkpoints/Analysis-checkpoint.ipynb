# Import libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings

warnings.filterwarnings('ignore')
sns.set()
```

# Data Preprocessing & EDA

```python
df = pd.read_csv('messages.csv')
```

```python
df.head()
```

## Feature extraction

```python
print(df['message'][0])
```

```python
floor_regex = r'Этаж: (\d+)'
area_regex = r'Площадь: (\d+) м²'
rooms_regex = r'Комнат: (\d+)'
bedrooms_regex = r'Спален: (\d+)'
district_regex = r'rent\. (.*)'
price_regex = r'Цена: (\d+)\$'
```

```python
df['district'] = df['message'].str.extract(district_regex)
df['area'] = df['message'].str.extract(area_regex)
df['floor'] = df['message'].str.extract(floor_regex)
df['rooms'] = df['message'].str.extract(rooms_regex)
df['bedrooms'] = df['message'].str.extract(bedrooms_regex)
df['price'] = df['message'].str.extract(price_regex)
```

```python
df.dropna(inplace=True)
```

```python
df['district'] = df['district'].str.strip()
df['area'] = df['area'].str.strip().astype(float)
df['floor'] = df['floor'].str.strip().astype(int)
df['rooms'] = df['rooms'].str.strip().astype(int)
df['bedrooms'] = df['bedrooms'].str.strip().astype(int)
df['price'] = df['price'].str.strip().astype(int)
```

```python
df['date'] = pd.to_datetime(df['date']).dt.normalize()
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(inplace=True)
df
```

```python
df.info()
```

## Analysis

```python
df['month'] = df['date'].astype("datetime64[M]")
df['week'] = df['date'].astype("datetime64[W]")
```

```python
plt.figure(figsize=(12, 9))

sns.lineplot(data=df, x='date', y='price');
```

```python
plt.figure(figsize=(12, 9))
for i in range(1, 4):
    data = df[df['rooms'] == i].groupby('date').rolling(14).mean().reset_index()
    sns.lineplot(data=data, x='date', y='price', label=f'{i} room(s)')
    
plt.title('Rolling 14 days price by rooms amount')
plt.show()
```

```python
plt.figure(figsize=(12, 9))
for i in range(1, 4):
    data = df[df['bedrooms'] == i].groupby('date').rolling(14).mean().reset_index()
    sns.lineplot(data=data, x='date', y='price', label=f'{i} room(s)')
    
plt.title('Rolling 14 days price by bedrooms amount')
plt.show()
```

```python
districts = list(df['district'].value_counts()[:5].index)

plt.figure(figsize=(12, 9))
for d in districts:
    data = df[df['district'] == d].groupby('date').rolling(14).mean().reset_index()
    sns.lineplot(data=data, x='date', y='price', label=f'{d}')

plt.title('Rolling 14 days price by top 5 District')
plt.show()
```
